## Defining Visualizations

```python
#Defining visualize must produce a sub plot of average of each features, max of each features and min of each features. 
def visualize(filename):
    "Plotting the mean, max and min of each dataset"

    data = numpy.loadtxt(fname=filename, delimiter=',')

    fig = matplotlib.pyplot.figure(figsize=(10.0, 3.0))

    axes1 = fig.add_subplot(1, 3, 1)

    axes1.set_ylabel('average')
    axes1.set_xlabel('Feature#')
    axes1.plot(numpy.mean(data, axis=0))
     
    axes1.set_ylabel('max')
    axes1.set_xlabel('Feature#')
    axes1.plot(numpy.max(data, axis=0))
    
    axes1.set_ylabel('min')
    axes1.set_xlabel('Feature#')
    axes1.plot(numpy.min(data, axis=0))    
    
    fig.tight_layout()
    matplotlib.pyplot.show()
```

## Defining Identify Issues

```python
#Defining identify issues, which prudces tatements based on certain conditions provided, see below
def identify_issues(filename):
    "Providing statements based on conditions"

    data = numpy.loadtxt(fname=filename, delimiter=',')

    if numpy.max(data, axis=0)[0] == 0 and numpy.max(data, axis=0)[20] == 20:
        print('I doubt the integrity of this dataset\n')
    elif numpy.sum(data.min())==0:
        print('Why is this happening')
    else:
        print('Dataset looks good')
```
